```python
#para este desafio primero importaremos pandas
import pandas as pd
```

```python
#ahora cargamos el documento donde tenemos los datos
df = pd.read_excel("resumen_resultados_astronautas.xlsx")
display(df)
```

```python
#necesitamos crear la fornula para calcular el IMC (imc=peso dividido por estatura al cuadrado
df['IMC'] = df['peso'] / ((df['altura'] / 100) ** 2)
display(df)
```

```python
#nos piden tener un filtro de IMC entre 18.5 y 24.9 asi que lo creamos
df = df[(df['IMC'] >= 18.5) & (df['IMC'] <= 24.9)]
display(df)
```

```python
# Paso 3: Calcular el promedio de las evaluaciones y crear la columna "promedio_evaluaciones"
df['promedio_evaluaciones'] = (df['evaluacion_fisica']+df['evaluacion_psicologica']+df['evaluacion_adaptativa']+df['evaluacion_medica'])/4

# Paso 4: Filtrar por promedio de evaluaciones mayor o igual a 87
df = df[df['promedio_evaluaciones'] >= 87]
```

```python
#reiniciamos el dataframe resultante
df = df.reset_index(drop=True)
display(df)
```

```python
#modificamos la columna 'califica' para que diga 'Si'
df['califica'] = 'Si'
display(df)
```

```python
# y para finalizar exportamos el DataFrame resultante a un archivo CSV
df.to_excel("astronautas_calificados.xlsx", index=False)
display(df)
```
